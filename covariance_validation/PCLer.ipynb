```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
import pymaster as nmt
import os
```

```python
nsims = 100
nside = 256
iband = 2
color = 'red'
npix = hp.nside2npix(nside)
lmax = 3*nside-1
ls = np.arange(lmax+1)
```

```python
mask = hp.ud_grade(hp.read_map("data/mask_apodized.fits"), nside_out=nside)
maskb = (mask > 0).astype(float)
```

```python
b = nmt.NmtBin.from_nside_linear(nside, nlb=10)
f = nmt.NmtField(mask, None, spin=2, purify_b=True, purify_e=True)
w = nmt.NmtWorkspace.from_fields(f, f, b)
leff = b.get_effective_ells()
```

```python
os.system('mkdir -p data/cls')
for i in range(nsims):
    seed = 1000 + i
    predir = 'data/cls'
    print(seed)
    q, u = hp.read_map(f'data/sims/sim_cmb_{seed}.fits', field=None)
    fs = nmt.NmtField(mask, [q*maskb, u*maskb], purify_b=True, purify_e=True)
    qn, un = hp.read_map(f'data/sims/sim_noise_{color}_band{iband}_{seed}.fits.gz',
                         field=None)
    fn = nmt.NmtField(mask, [qn*maskb, un*maskb], purify_b=True, purify_e=True)

    fname = f'{predir}/cls_sig_{seed}.npz'
    pcl = nmt.compute_coupled_cell(fs, fs)
    cl = w.decouple_cell(pcl)
    np.savez(fname, cl=cl, pcl=pcl)

    fname = f'{predir}/cls_signoi_{color}_band{iband}_{seed}.npz'
    pcl = nmt.compute_coupled_cell(fs, fn)
    cl = w.decouple_cell(pcl)
    np.savez(fname, cl=cl, pcl=pcl)

    fname = f'{predir}/cls_noi_{color}_band{iband}_{seed}.npz'
    pcl = nmt.compute_coupled_cell(fn, fn)
    cl = w.decouple_cell(pcl)
    np.savez(fname, cl=cl, pcl=pcl)
```
